```python
import numpy as np
import pandas as pd
import sklearn, warnings
from mlwpy import *
%matplotlib inline
```

```python
data = pd.read_csv('flaringar_original_data.csv')
data
```

```python
#example
#input: X22
#returns X

def get_class(exp):
    return exp[0]
```

```python
#reformatting the 'Flare Class' column in the dataset

flare_tgt = data['Flare Class'].apply(get_class)
flare_tgt
```

```python
#dropping 2 columns
data = data.drop(['Flare Class', 'Flare Date', 'AR'], axis=1)
```

```python
warnings.filterwarnings('ignore',
                        category=FutureWarning,
                        module='sklearn')

warnings.filterwarnings('ignore',
                        category=FutureWarning,
                        module='tensor*')
```

```python
(flare_train_ftrs, flare_test_ftrs,
flare_train_tgt, flare_test_tgt) = skms.train_test_split(data,
                                                       flare_tgt,
                                                       test_size=.25)
print("Train features shape:", flare_train_ftrs.shape)
print("Test features shape:", flare_test_ftrs.shape)
```

```python
#3-nn classification model
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
fit = knn.fit(flare_train_ftrs, flare_train_tgt)
preds = fit.predict(flare_test_ftrs)

metrics.accuracy_score(flare_test_tgt, preds)
```
